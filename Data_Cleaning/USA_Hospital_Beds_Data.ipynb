```python
import pandas as pd
```

```python
df = pd.read_json("../Data/Raw Data/usa-hospital-beds_geo.json", lines= True)
df.head()
```

```python
df = df.drop(['HQ_ADDRESS', 'HQ_ADDRESS1', 'HQ_CITY', 'HQ_STATE', 'HQ_ZIP_CODE', 'latitude', 'longtitude'], axis= 1)
```

```python
df['NUM_LICENSED_BEDS'] = df['NUM_LICENSED_BEDS'].fillna(0)
df['NUM_LICENSED_BEDS'] = df['NUM_LICENSED_BEDS'].astype(int)
df['NUM_LICENSED_BEDS']
```

```python
df['NUM_STAFFED_BEDS'] = df['NUM_STAFFED_BEDS'].fillna(0)
df['NUM_STAFFED_BEDS'] = df['NUM_STAFFED_BEDS'].astype(int)
df['NUM_STAFFED_BEDS']
```

```python
df['PEDI_ICU_BEDS'] = df['PEDI_ICU_BEDS'].fillna(0)
df['PEDI_ICU_BEDS'] = df['PEDI_ICU_BEDS'].astype(int)
df['PEDI_ICU_BEDS']
```

```python
df['BED_UTILIZATION'] = df['BED_UTILIZATION'].fillna(0)
df['BED_UTILIZATION'] = df['BED_UTILIZATION'].astype(float)
df['BED_UTILIZATION']
```

```python
df['AVG_VENTILATOR_USAGE'] = df['AVG_VENTILATOR_USAGE'].fillna(0)
df['AVG_VENTILATOR_USAGE'] = df['AVG_VENTILATOR_USAGE'].astype(int)
df['AVG_VENTILATOR_USAGE']
```

```python
df.head()
```

```python
df.to_csv("../Data/Clean Data/USA_Hospital_Beds.csv", index= False)
```
