```python
import pandas as pd
```

```python
df = pd.read_csv("../Data/Raw Data/us_daily.csv")
df.head()
```

```python
df = df.drop(['dateChecked', 'lastModified', 'recovered', 'total', 'posNeg', 'hash'], axis= 1)
df.tail()
```

```python
columns = df.columns
for i in columns:
    df[i] = df[i].fillna(0)   
    df[i] = df[i].astype(int)
    df[i]
```

```python
Date = []
for i in df['date']:
    date_object = pd.to_datetime(i, format='%Y%m%d')
    formatted_date = date_object.strftime('%Y-%m-%d')
    Date.append(formatted_date)
df['date'] = Date
```

```python
df.head()
```

```python
df.to_csv("../Data/Clean Data/US_Daily_Cases.csv", index= False)
```
