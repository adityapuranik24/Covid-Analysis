### Loading Country Codes DataSet

```python
country_codes = pd.read_csv("../../Data/Clean Data/Updated_Country_Codes.csv")
country_codes
```

```python
# Create Table
conn.execute('''
		CREATE TABLE country_codes (
			iso_code nvarchar(10) PRIMARY KEY,
			continent nvarchar(20),
			location nvarchar(40)
			)
               ''')

# Insert DataFrame to Table
for row in country_codes.itertuples():
    conn.execute('''
                INSERT INTO country_codes (iso_code, continent, location)
                VALUES (?,?,?)
                ''',
                row.iso_code, 
                row.continent,
                row.location
                )
conn.commit()
```

### Loading World Population DataSet 

```python
population = pd.read_csv("../../Data/Clean Data/world_population.csv")
population
```

```python
# Create Table
conn.execute('''
		CREATE TABLE world_population (
			iso_code nvarchar(10),
			population int,
			population_density float
            FOREIGN KEY (iso_code) REFERENCES country_codes(iso_code)
			)
               ''')

# Insert DataFrame to Table
for row in population.itertuples():
    conn.execute('''
                INSERT INTO world_population (iso_code, population, population_density)
                VALUES (?,?,?)
                ''',
                row.iso_code, 
                row.population,
                row.population_density
                )
conn.commit()
```
